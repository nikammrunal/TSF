```python
# importing all the required Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Loading the Dataset
data = pd.read_csv('../Datasets/SampleSuperstore.csv')
data.head()
```

```python
data.shape
```

### Exploring the Dataset

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Country'],axis=1,inplace=True)
data
```

### Data Visualization

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(70,70))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per sub-Category')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
data['Category'].value_counts().plot.bar(figsize=(20,20),color='blue',title='Category')
plt.subplot(222)
data['Ship Mode'].value_counts().plot.bar(figsize=(20,20),color='red',title='Ship Mode')
plt.subplot(223)
data['Region'].value_counts().plot.bar(figsize=(20,20),color='grey',title='Region')
plt.subplot(224)
data['Segment'].value_counts().plot.bar(figsize=(20,20),color='green',title='Segment')
plt.show()
```

so by firts figure it shows that, most of the category are Supplies, and most of the ship mode are of standard class, region is distributed but maximum  is from west region and mostly Segment are Consumer.

```python
plt.figure(figsize=(16,16))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
sns.scatterplot("Sales",'Profit',data=data)
```

```python
pd.crosstab(data["Quantity"],data["Category"],data["Profit"],aggfunc='sum').plot(kind="bar",stacked=True)
```

```python
sns.catplot(x='Sub-Category',y="Profit",data= data,color='Green',height=4,aspect=4)
```

from this catelog, it can be viewed Blinders are highly variable and Copiers earns maximum Profit.

```python
data.plot(kind='scatter',x='Quantity', y='Profit', figsize=(12,6),color='red')
plt.show()
```

form this scatter plot , it is found that Profit is also not depending based upon Quantity.There are many other factors acting upon it.
